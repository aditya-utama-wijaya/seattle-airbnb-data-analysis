# Data preparation

Before we dive into the analysis, let's first prepare our data. We'll start by importing the necessary libraries and loading the dataset:

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

listing = pd.read_csv('listings.csv')
print(listing.shape)
pd.concat([listing.head(), listing.tail()])
```

We'll also clean the data by converting the 'price' column to an integer value:

```python
listing['price']
```

```python
listing["price"] = listing["price"].map(lambda p: int(p[1:-3].replace(",", "")))
listing["price"]
```

# Analysis

### 1. What are the most and least expensive neighborhoods to stay in Seattle?

Let's start by exploring the most expensive neighborhoods in Seattle based on Airbnb prices. We'll use a boxplot to visualize the distribution of prices for each neighborhood:

```python
sort_order = (listing
              .groupby('neighbourhood_cleansed')['price']
              .median()
              .sort_values(ascending = False)
              .index
             )
plt.figure(figsize = (24, 12))
sns.boxplot(y = 'price', x = 'neighbourhood_cleansed', data = listing, order = sort_order)
plt.xticks(rotation = 45, ha = 'right')
plt.xlabel('Neighborhood', fontsize = 14)
plt.ylabel('Price ($)', fontsize = 14)
plt.title('Distribution of Airbnb Prices by Neighborhood in Seattle\n', fontsize = 20)
```

We can see that the most expensive neighborhoods in Seattle are `Briarcliff`, `Pioneer Square`, and `Westlake`. Meanwhile the least expensive neighborhoods in Seattle are `South Beacon Hill`, `Rainier Beach`, and `Olympic Hills`.

### 2. How do listing prices vary by property type and room type?

```python
sort_order = (listing
              .groupby('property_type')['price']
              .median()
              .sort_values(ascending = False)
              .index
             )
plt.figure(figsize = (12, 8))
sns.boxplot(y = 'price', x = 'property_type', data = listing, order = sort_order)
plt.xticks(rotation = 45, ha = 'right')
```

In terms of `property_type`, we can see that `Boat` has the highest distribution of prices, and `Dorm` has the lowest one.

```python
sort_order = (listing
              .groupby('room_type')['price']
              .median()
              .sort_values(ascending = False)
              .index
             )
plt.figure(figsize = (12, 8))
sns.boxplot(y = 'price', x = 'room_type', data = listing, order = sort_order)
```

In terms of `room_type`, we can see that `Entire home/apt` has the highest distribution of prices, and `Shared room` has the lowest one.

```python
sort_order = (listing
              .groupby('bed_type')['price']
              .median()
              .sort_values(ascending = False)
              .index
             )
plt.figure(figsize = (12, 8))
sns.boxplot(y = 'price', x = 'bed_type', data = listing, order = sort_order)
```

In terms of `bed_type`, we can see that `Real Bed` has the highest distribution of prices, and `Couch` has the lowest one.

### 3. Which amenities are associated with higher listing prices?

First, let's explore the most popular amenities in Seattle Airbnb listings. We'll start by creating a list of all amenities and their frequencies:

```python
listing['amenities'] = (listing['amenities']
                        .map(lambda am: '|'.join([a.replace('}', '').replace('{', '').replace('"', '') for a in am.split(',')]))
                       )
amenities = pd.Series(np.concatenate(listing['amenities'].map(lambda a: a.split('|'))))
amenities = amenities[amenities != '']
amenities
```

We can then plot the top 20 amenities by frequency:

```python
plt.figure(figsize = (12, 8))
amenities.value_counts().nlargest(20).plot(kind = 'bar')
plt.xticks(rotation = 45, ha = 'right')
plt.ylabel('Count\n', fontsize = 14)
plt.xlabel('\nAmenities', fontsize = 14)
plt.title('Top 20 Most Popular Amenities in Seattle Airbnb Listings\n', fontsize = 20)
```

We can see that the most popular amenities in Seattle Airbnb listings are `Wireless Internet`, `Heating`, and `Kitchen`.

```python
amenities = np.unique(amenities)
amenity_prices = [(a, listing[listing['amenities'].map(lambda am: a in am)]['price'].mean()) for a in amenities]
amenity_prices
```

```python
amenity_srs = pd.Series(data = [a[1] for a in amenity_prices], index = [a[0] for a in amenity_prices])
amenity_srs
```

```python
plt.figure(figsize = (12, 8))
amenity_srs.sort_values(ascending = False).plot(kind = 'bar')
plt.xticks(rotation = 45, ha = 'right')
plt.xlabel('\nAmenities', fontsize = 14)
plt.ylabel('Mean Price ($)\n', fontsize = 14)
plt.title('Distribution of Listing Price by Amenities\n', fontsize = 20)
```

We can see that the most common amenities in expensive listings are `Doorman`, `Pool`, and `Hot Tub`.
